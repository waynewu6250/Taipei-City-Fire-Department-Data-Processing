```python
data = pd.read_excel("1.xlsx")
data
```

```python
def insert(m,tmp):
    if m != None:
        tmp.append(m.group(1))
    else:
        tmp.append(0)
    return tmp

def f(x, match):
    if x in match:
        return match[x]
    else: return x
```

```python
match = {'一': 1, '二': 2,'三': 3,'四': 4,'五': 5, '六': 6,'七': 7,'八': 8,'九': 9,'十': 10,
         '十一': 11,'十二': 12,'十三': 13,'十四': 14,'十五': 15,'十六': 16,'十七': 17,'十八': 18,'十九': 19,'二十': 20}
address = data["居住地址：街路門牌"]
data["路"] = address.map(lambda x: re.match('(.*?)\d+',x).group(1))

sec = []
lane = []
sub_lane = []
neighbor = []
number = []
sub_number = []
floor = []
sub_floor = []

for i in range(len(address)):
    m1 = re.match('.*[路街](\d*)段',address.iloc[i])
    m2 = re.match('.*[街路段](\d*)巷',address.iloc[i])
    m3 = re.match('.*巷(\d*)弄',address.iloc[i])
    m4 = re.match('.*弄(\d*)鄰',address.iloc[i])
    m5 = re.match('.*[街路段巷弄鄰](\d*)[之號]',address.iloc[i])
    m6 = re.match('.*之(\d*)號',address.iloc[i])
    m7 = re.match('.*號(\w*)樓',address.iloc[i])
    m8 = re.match('(^[0-9]*$)',address.iloc[i].split("之")[-1])
    
    sec = insert(m1,sec)
    lane = insert(m2, lane)
    sub_lane = insert(m3, sub_lane)
    neighbor = insert(m4, neighbor)
    number = insert(m5, number)
    sub_number = insert(m6, sub_number)
    floor = insert(m7, floor)
    sub_floor = insert(m8, sub_floor)
    
        
data["段"] = np.array(sec)
data["巷"] = np.array(lane)
data["弄"] = np.array(sub_lane)
data["鄰"] = np.array(neighbor)
data["號"] = np.array(number)
data["號之"] = np.array(sub_number)
data["樓"] = np.array(floor)
data["樓之"] = np.array(sub_floor)
data["樓"] = data["樓"].apply(lambda x: f(x,match))
data
```

```python
# Parse the number
mindata = data[["聯絡電話(住家)", "聯絡電話(辦公)", "行動電話"]]
telephone = []
mobile = []
contact_tel = []
contact_mob = []

for i in range(len(mindata)):
    a1 = re.match('(^2\d{7})(\d*)', str(mindata.iloc[i]["聯絡電話(住家)"]))
    a2 = re.match('(^2\d{7})(\d*)', str(mindata.iloc[i]["聯絡電話(辦公)"]))
    a3 = re.match('(^2\d{7})(\d*)', str(mindata.iloc[i]["行動電話"]))
    b1 = re.match('(^9\d{8})(\d*)', str(mindata.iloc[i]["聯絡電話(住家)"]))
    b2 = re.match('(^9\d{8})(\d*)', str(mindata.iloc[i]["聯絡電話(辦公)"]))
    b3 = re.match('(^9\d{8})(\d*)', str(mindata.iloc[i]["行動電話"]))
    c1 = re.match('(^2\d{7})(\d*)', str(data.iloc[i][12]))
    c2 = re.match('(^2\d{7})(\d*)', str(data.iloc[i][13]))
    d1 = re.match('(^9\d{8})(\d*)', str(data.iloc[i][12]))
    d2 = re.match('(^9\d{8})(\d*)', str(data.iloc[i][13]))
    
    if a1 != None:
        telephone.append("02"+a1.group(1))
    elif a2 != None:
        telephone.append("02"+a2.group(1))
    elif a3 != None:
        telephone.append("02"+a3.group(1))
    else:
        telephone.append("0")
    
    if b1 != None:
        mobile.append("0"+b1.group(1))
    elif b2 != None:
        mobile.append("0"+b2.group(1))
    elif b3 != None:
        mobile.append("0"+b3.group(1))
    else:
        mobile.append("0")
        
    if c1 != None:
        contact_tel.append("02"+c1.group(1))
    elif c2 != None:
        contact_tel.append("02"+c2.group(1))
    else:
        contact_tel.append("0")
    
    if d1 != None:
        contact_mob.append("0"+d1.group(1))
    elif d2 != None:
        contact_mob.append("0"+d2.group(1))
    else:
        contact_mob.append("0")

data["telephone"] = np.array(telephone)
data["mobile"] = np.array(mobile)
data["contact_tel"] = np.array(contact_tel)
data["contact_mob"] = np.array(contact_mob)
```
