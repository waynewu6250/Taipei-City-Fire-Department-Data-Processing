```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
def weather_check(x):
    if x == '...':
        return ''
    x = float(x)
    if x >= 25:
        return "熱"
    elif x >= 20:
        return "溫暖"
    elif x >= 10:
        return "涼"
    elif x < 10:
        return "冷"
def rain_check(x):
    if x == '...':
        return ''
    x = float(x)
    if x >= 10:
        return "雨天"
    elif x > 0: 
        return "飄雨"
    else: return "沒雨"
def sunshine_check(x):
    if x == '...':
        return ''
    x = float(x)
    if x >= 20:
        return "艷陽"
    elif x >= 10:
        return "晴天"
    elif x >= 0:
        return "陰天"
    else: return ""
```

```python
all_data = []

for i in range(1,4):
    df = pd.read_csv('466920-2019-0{}.csv'.format(i))
    
    data = df[["Temperature", "Precp", "PrecpHour", "SunShineRate"]]
    #data["Weather"] = data["Temperature"].apply(lambda x: weather_check(x))
    data["Rain"] = data["PrecpHour"].apply(lambda x: rain_check(x))
    data["Sunshine"] = data["SunShineRate"].apply(lambda x: sunshine_check(x))
    
    data["Overall"] = data["Sunshine"].values+" "+data["Rain"].values+" "+data["Temperature"].apply(str)+" C"
    all_data.append(data[["Temperature", "Rain", "Overall"]])
```

```python
len(all_data)
```

```python
d = pd.concat(all_data, axis=1)
d.to_excel("ouptut.xlsx")
```

# Analysis

```python
xlsx = pd.ExcelFile('本局救護未送醫案件分析表0314.xlsx')
df1 = pd.read_excel(xlsx, '108年1月').reset_index()
df1["weather"] = d.iloc[:,0]
df2 = pd.read_excel(xlsx, '108年2月').reset_index()
df2["weather"] = d.iloc[:,2]
df3 = pd.read_excel(xlsx, '108年3月').reset_index()
df3["weather"] = d.iloc[:,4]
```

```python
df = pd.concat([df1,df2,df3], axis=0)
```

```python
df = df[['出勤次數', '拒送', 'weather']].dropna().reset_index()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df["weather"])
df["weather_count"] = le.transform(df["weather"])
print(list(le.classes_))
```

```python
x = np.asarray(range(1,len(df)+1))
y = df["拒送"].values
label = df["weather_count"].values
colors = ['red','green','blue']
fig = plt.figure(figsize=(16,8))
plt.scatter(x, y, c=label, cmap=matplotlib.colors.ListedColormap(colors))
plt.xlabel('day', size='20')
plt.ylabel('times', size='20')

cb = plt.colorbar()
loc = np.arange(0,max(label),max(label)/float(len(colors)))
cb.set_ticks(loc)
cb.set_ticklabels(['sunny', 'heavy rain', 'rainy'])

plt.savefig('天氣與拒送次數關係圖.jpg')

print("沒雨的平均拒送次數",np.mean(df[df["weather"]=="沒雨"]["拒送"].values))
print("雨天的平均拒送次數",np.mean(df[df["weather"]=="雨天"]["拒送"].values))
print("飄雨的平均拒送次數",np.mean(df[df["weather"]=="飄雨"]["拒送"].values),'\n')

print("沒雨的平均出勤次數",np.mean(df[df["weather"]=="沒雨"]["出勤次數"].values))
print("雨天的平均出勤次數",np.mean(df[df["weather"]=="雨天"]["出勤次數"].values))
print("飄雨的平均出勤次數",np.mean(df[df["weather"]=="飄雨"]["出勤次數"].values))
```
