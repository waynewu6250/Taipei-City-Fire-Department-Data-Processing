# Satisfaction records patch

```python
browser = open_browser()
browser = login(browser)
browser = find_page(browser)
```

```python
data = []

while True:
    
    newdata = read_data(browser)
    
    browser, final, numbers = insert_all_data(browser, newdata)
    final["number"] = np.asarray(numbers[::-1])
    data.append(final)
    all_data = data[len(data)-1]
    if len(all_data) == 50:
        break
    
    nextbutton = browser.find_element_by_name("gonext")
    if nextbutton.get_attribute("onclick") == None:
        break
    else:
        nextbutton.click()
```

```python
alldata = data[len(data)-1]
if len(data) > 1:
    for i in range(len(data)-2,-1,-1):
        alldata = alldata.append(data[i], ignore_index=True)

alldata = alldata[alldata['number'].str.contains('^09') == True]
alldata.iloc[:,2] = alldata.iloc[:,2].apply(lambda x: ":".join(x.replace("-","/").split(":")[:2]))
alldata.to_excel("output.xlsx")
os.startfile("output.xlsx")
```
